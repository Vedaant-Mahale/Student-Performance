```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Student_Performance.csv')
df.loc[df['Extracurricular Activities'] == 'Yes','Extracurricular Activities'] = 1 # Making all Yes Values 1s
df.loc[df['Extracurricular Activities'] == 'No','Extracurricular Activities'] = 0 # Making all No Values 0s
df = df.corr()
sns.heatmap(df, annot=True, cmap='coolwarm')
plt.show() # High Correlation between Hours Studied and Performance, as well as with Previous scores. Low Relations with all other combinations
```

```python
df = pd.read_csv('Student_Performance.csv')
df = df[df["Performance Index"] >= 80] # Finding Heatmap for All students who have a performance index above or equal to 80
df.loc[df['Extracurricular Activities'] == 'Yes','Extracurricular Activities'] = 1 # Making all Yes Values 1s
df.loc[df['Extracurricular Activities'] == 'No','Extracurricular Activities'] = 0 # Making all No Values 0s
df = df.corr()
sns.heatmap(df, annot=True, cmap='coolwarm')
plt.show() # Previous Scores and Hours studied actually become inversely related and both are equally affecting the performance index
```

```python
df = pd.read_csv('Student_Performance.csv')
df = df[(df["Performance Index"] < 80) & (df["Performance Index"] > 50)] # Finding Heatmap for All students who have a performance index between 50 and 80
df.loc[df['Extracurricular Activities'] == 'Yes','Extracurricular Activities'] = 1 # Making all Yes Values 1s
df.loc[df['Extracurricular Activities'] == 'No','Extracurricular Activities'] = 0 # Making all No Values 0s
df = df.corr()
sns.heatmap(df, annot=True, cmap='coolwarm')
plt.show() # Almost No correlation between Performance and other variables besides previous scores. Previous scores and Hours studied are inversely related
```

```python
df = pd.read_csv('Student_Performance.csv')
df = df[df["Performance Index"] < 50] # Finding Heatmap for All students who have a performance index less than 50
df.loc[df['Extracurricular Activities'] == 'Yes','Extracurricular Activities'] = 1 # Making all Yes Values 1s
df.loc[df['Extracurricular Activities'] == 'No','Extracurricular Activities'] = 0 # Making all No Values 0s
df = df.corr()
sns.heatmap(df, annot=True, cmap='coolwarm')
plt.show() # Performance index is affected by hours studied and previous scores. Previous scores and Hours studied are inversely related
```
